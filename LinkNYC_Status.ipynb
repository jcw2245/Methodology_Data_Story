```python
duplicates = df.duplicated(subset=['ADDRESS'], keep=False)
duplicate_rows = df[duplicates]

print(duplicate_rows)
```

```python
address_dataset = duplicate_rows[['ADDRESS']].copy()

address_dataset.loc[:, 'CITY'] = duplicate_rows['CITY']

address_dataset.loc[:, 'STATUS'] = duplicate_rows['STATUS']

address_dataset.to_csv('address_with_city_and_status_dataset.csv', index=False)

```

```python

address_dataset = duplicate_rows[['ADDRESS', 'CITY', 'STATUS', 'GENERATED ON']].copy()


address_dataset.to_csv('address_with_city_status_and_generated_on.csv', index=False)

```
