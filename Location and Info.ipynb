```python
df['Installation Complete'] = pd.to_datetime(df['Installation Complete'], format='%m/%d/%Y')
df['Activation Complete'] = pd.to_datetime(df['Activation Complete'], format='%m/%d/%Y')


df['Time Difference'] = (df['Activation Complete'] - df['Installation Complete']).dt.days

average_time_difference = df['Time Difference'].mean()
average_time_difference
```
